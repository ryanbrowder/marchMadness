```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_actualsSmoother = pd.read_csv('data/actualsSmoother_L0.csv')

display(df_actualsSmoother.head(5))
display(df_actualsSmoother.info())
```

```python
df_actualsSmoother.columns = df_actualsSmoother.iloc[0]  # Set the first row as column names
df_actualsSmoother = df_actualsSmoother[1:].reset_index(drop=True)  # Remove the first row and reset index

df_actualsSmoother[['Seed', 'Index', 'Actual']] = df_actualsSmoother[['Seed', 'Index', 'Actual']].apply(pd.to_numeric, errors='coerce').astype('Int64')
df_actualsSmoother['Projected'] = pd.to_numeric(df_actualsSmoother['Projected'], errors='coerce')

display(df_actualsSmoother.head(5))
display(df_actualsSmoother.info())
```

```python
# Convert to csv

df_actualsSmoother.to_csv("../L1/data/actualsSmoother_L1.csv")
```
