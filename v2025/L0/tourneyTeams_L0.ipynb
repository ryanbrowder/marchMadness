```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_tourneyTeams = pd.read_csv('data/tourneyTeams_L0.csv')

df_teamsIndex = pd.read_csv('../L1/data/teamsIndex_L1.csv')

display(df_tourneyTeams.head(5))
display(df_tourneyTeams.info())

df_tourneyTeams.to_csv("../L1/data/calcuttaNATOR_L1.csv")
```

```python
df_tourneyTeams = df_tourneyTeams.drop(columns=['Golden Ratio', 'VEGsemble'])
display(df_tourneyTeams.head(5))
display(df_tourneyTeams.info())
```

```python
# Perform a left join on "Team"
df_tourneyTeams = df_tourneyTeams.merge(df_teamsIndex, on='Team', how='left')

df_tourneyTeams = df_tourneyTeams.drop(columns=['Unnamed: 0'])

# Display the first few rows to verify
print(df_tourneyTeams.head())
```

```python
# Filter rows where "Index" is NaN in df_lrmcC
df_tourneyTeams_nan_index = df_tourneyTeams[df_tourneyTeams['Index'].isna()]

# Display the filtered DataFrame
print(df_tourneyTeams_nan_index)
```
