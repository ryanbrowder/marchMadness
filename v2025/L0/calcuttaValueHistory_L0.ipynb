```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_calcuttaValueHistory = pd.read_csv('data/calcuttaValueHistory_L0.csv')

display(df_calcuttaValueHistory.head(5))
display(df_calcuttaValueHistory.info())
```

```python
# List of columns that should remain as integers
int_columns = ['Unnamed: 0', 'Entrants', 'Winning Pts', 'Total Points']

# Convert all other columns to float after removing any dollar signs or commas
for col in df_calcuttaValueHistory.columns:
    if col not in int_columns:
        df_calcuttaValueHistory[col] = df_calcuttaValueHistory[col].replace('[\$,]', '', regex=True).astype(float)


df_calcuttaValueHistory.rename(columns={'Unnamed: 0': 'Year'}, inplace=True)

display(df_calcuttaValueHistory.head(25))
display(df_calcuttaValueHistory.info())
```

```python
# Convert to csv

df_calcuttaValueHistory.to_csv("../L1/data/calcuttaValueHistory_L1.csv")
```
