```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_kenPom = pd.read_csv('data/kenPom_L0.csv')

df_teamsIndex = pd.read_csv('../L1/data/teamsIndex_L1.csv')

display(df_kenPom.head(5))
display(df_kenPom.info())
```

```python
# Rename column
df_kenPom.rename(columns={'Rk': 'KenPom Rank'}, inplace=True)

# Keep only the relevant columns
df_kenPom = df_kenPom[['Team', 'KenPom Rank']]

# Drop any non-numeric rows (e.g., header row mistakenly included)
df_kenPom = df_kenPom[df_kenPom['KenPom Rank'].str.match(r'^\d+(\.\d+)?$', na=False)]

# Strip spaces and convert "KenPom Rank" to float
df_kenPom['KenPom Rank'] = df_kenPom['KenPom Rank'].astype(str).str.strip().astype(float)

import re

# Remove any numeric characters from "Team"
df_kenPom['Team'] = df_kenPom['Team'].apply(lambda x: re.sub(r'\d+', '', x).strip())

# Display results
display(df_kenPom.head(5))
display(df_kenPom.info())
```

```python
# Perform a left join on "Team"
df_kenPom = df_kenPom.merge(df_teamsIndex, on='Team', how='left')

df_kenPom = df_kenPom.drop(columns=['Unnamed: 0'])

# Display the first few rows to verify
print(df_kenPom.head(5))
```

```python
# Convert to csv

df_kenPom.to_csv("../L1/data/kenPom_L1.csv")
```
