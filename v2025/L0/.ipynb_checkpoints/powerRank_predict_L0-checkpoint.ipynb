```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_powerRank = pd.read_csv('data/powerRank_L0.csv')

df_teamsIndex = pd.read_csv('../L1/data/teamsIndex_L1.csv')

display(df_powerRank.head(5))
display(df_powerRank.info())
```

```python
df_powerRank[['Team', 'Record']] = df_powerRank['Team'].str.extract(r'(.+?)\s*\((\d+-\d+)\)')

df_powerRank = df_powerRank.drop(columns=['Rating', 'Record'])

df_powerRank.rename(columns={'Rank': 'PowerRank Rank'}, inplace=True)

display(df_powerRank.head(5))
display(df_powerRank.info())
```

```python
# Perform a left join on "Team"
df_powerRank = df_powerRank.merge(df_teamsIndex, on='Team', how='left')

df_powerRank = df_powerRank.drop(columns=['Unnamed: 0'])

# Display the first few rows to verify
print(df_powerRank.head(5))
```

```python
# Convert to csv

df_powerRank.to_csv("../L1/data/powerRank_L1.csv")
```
