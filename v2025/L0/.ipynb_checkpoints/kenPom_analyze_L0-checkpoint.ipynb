```python
#analyze file
#grabbed kenPom data from 2014-24 (no 20 / covid)
#looking for correlations in column fields that would predict the elite 8 flag
```

```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_kenPom = pd.read_csv('data/kenPom_analyze_L0.csv')
df_bluePower = pd.read_csv('data/bluePower_L0.csv')

display(df_kenPom.head(5))
display(df_kenPom.info())

display(df_bluePower.head(5))
display(df_bluePower.info())
```

```python
#drop columns

df_kenPom = df_kenPom.drop(columns=["Rk","L","r","r.1","r.2","r.3","r.4","r.5","r.6","r.7"], errors="ignore")
```

```python
#for the column "Elite 8", any row value that is NaN should be a 0 and sets all types to int

df_kenPom["Elite 8"] = df_kenPom["Elite 8"].fillna(0).astype(int)
```

```python
#marks non tourney teams with a seed of 100

# Remove spaces and any non-numeric characters (except for possible decimal points)
df_kenPom["Seed"] = df_kenPom["Seed"].astype(str).str.replace(r"[^\d.]", "", regex=True)
```

```python
import re

df_kenPom['Team'] = df_kenPom['Team'].apply(lambda x: re.sub(r'\d+$', '', x).strip() if isinstance(x, str) else x)

display(df_kenPom.head(5))
display(df_kenPom.info())
```

```python
df_kenPom = df_kenPom.merge(df_bluePower, on="Team", how="left")

df_kenPom = df_kenPom.drop(columns=["Tournament_Appearances"])  # Drop column
df_kenPom = df_kenPom.rename(columns={"Fibonacci_Score": "bluePower"})  # Rename column

display(df_kenPom.head(5))
display(df_kenPom.info())
```

```python
#columns to float using astype(float

numbers = ["Seed","W","NetRtg","ORtg","DRtg","AdjT","Luck","SoSNetRtg","SoSORtg","SoSDRtg","ConfSoSNetRtg","bluePower"]

df_kenPom[numbers] = df_kenPom[numbers].apply(pd.to_numeric, errors="coerce")
```

```python
# Replace NaN Seed teams; these didn't make the tournament

df_kenPom = df_kenPom[df_kenPom["Seed"].notna() & (df_kenPom["Seed"] > 0)]

display(df_kenPom.head(5))
display(df_kenPom.info())
```

```python
#invert the ranks/seeds, so that higher values are transformed into lower ones, aligning better with their actual value.

#df_kenPom["Rk"] = df_kenPom["Rk"].max() - df_kenPom["Rk"] + 1
df_kenPom["Seed"] = df_kenPom["Seed"].max() - df_kenPom["Seed"] + 1
```

```python
# wins of 0 should be turned to 12

df_kenPom["W"] = df_kenPom["W"].replace("", 12).fillna(12)
```

```python
# "Conf" column in df_torvik by replacing 'ind' with 'Ind'

df_kenPom["Conf"] = df_kenPom["Conf"].str.replace(r"^ind$", "Ind", case=False, regex=True)
```

```python
# one-hot encode the "Conf" column

df_kenPom = pd.get_dummies(df_kenPom, columns=["Conf"], drop_first=False)
```

```python
#new column named "AdjJOD" at index #7. The value is 1 if "AdjOE" is > 119 and if "AdjDE" is <94. Else the value is 0.

df_kenPom.insert(7, "AdjOD", np.where((df_kenPom["ORtg"] > 116) & (df_kenPom["DRtg"] < 96), 1, 0))

display(df_kenPom.head(5))
display(df_kenPom.info())
```

```python
#swap T/F with 1/0

CONF = ["Conf_A10","Conf_ACC","Conf_AE","Conf_ASun","Conf_Amer","Conf_B10","Conf_B12","Conf_BE","Conf_BSky","Conf_BSth","Conf_BW","Conf_CAA","Conf_CUSA","Conf_Horz","Conf_Ivy","Conf_MAAC","Conf_MAC","Conf_MEAC","Conf_MVC","Conf_MWC","Conf_NEC","Conf_OVC","Conf_P12","Conf_Pat","Conf_SB","Conf_SC","Conf_SEC","Conf_SWAC","Conf_Slnd","Conf_Sum","Conf_WAC","Conf_WCC"]

df_kenPom[CONF] = df_kenPom[CONF].astype(int)
```

```python
#To apply a log transformation to columns

df_kenPom[numbers] = np.log(df_kenPom[numbers] + 1)
```

```python
#replace NaN values in the specified columns with 0 

df_kenPom[["NetRtg", "SoSNetRtg", "ConfSoSNetRtg"]] = df_kenPom[["NetRtg", "SoSNetRtg", "ConfSoSNetRtg"]].fillna(0)
```

```python
display(df_kenPom.head(5))
display(df_kenPom.info())
```

```python
df_kenPom['bluePower_NetRtg'] = df_kenPom['bluePower'] * df_kenPom['NetRtg']
df_kenPom['bluePower_Seed'] = df_kenPom['bluePower'] * df_kenPom['Seed']

# Display the updated df_torvik with the new interaction feature
display(df_kenPom[['Team', 'bluePower', 'NetRtg', 'bluePower_NetRtg']].head())
display(df_kenPom[['Team', 'bluePower', 'Seed', 'bluePower_Seed']].head())
```

```python
# Convert to csv

df_kenPom.to_csv("../L1/data/kenPom_analyze_L1.csv")
```
