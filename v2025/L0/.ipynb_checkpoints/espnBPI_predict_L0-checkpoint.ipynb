```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_espnBPI = pd.read_csv('data/espnBPI_L0.csv')

df_teamsIndex = pd.read_csv('../L1/data/teamsIndex_L1.csv')

display(df_espnBPI.head(5))
display(df_espnBPI.info())
```

```python
df_espnBPI.rename(columns={'Unnamed: 0': 'Team', 'Unnamed: 5': 'BPI Rank'}, inplace=True)

df_espnBPI = df_espnBPI[['Team', 'BPI Rank']]

# Drop any non-numeric rows (e.g., header row mistakenly included)
df_espnBPI = df_espnBPI[df_espnBPI['BPI Rank'].str.match(r'^\d+(\.\d+)?$', na=False)]

# Strip spaces and convert "KenPom Rank" to float
df_espnBPI['BPI Rank'] = df_espnBPI['BPI Rank'].astype(str).str.strip().astype(float)

df_espnBPI = df_espnBPI.dropna()

display(df_espnBPI.head(5))
display(df_espnBPI.info())
```

```python
# Perform a left join on "Team"
df_espnBPI = df_espnBPI.merge(df_teamsIndex, on='Team', how='left')

df_espnBPI = df_espnBPI.drop(columns=['Unnamed: 0'])

# Display the first few rows to verify
print(df_espnBPI.head(5))
```

```python
# Convert to csv

df_espnBPI.to_csv("../L1/data/espnBPI_L1.csv")
```
