```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_lrmcB = pd.read_csv('data/lrmcB_predict_L0.csv')

df_teamsIndex = pd.read_csv('../L1/data/teamsIndex_L1.csv')

display(df_lrmcB.head(5))
display(df_lrmcB.info())
```

```python
df_lrmcB.rename(columns={'Unnamed: 1': 'LRMC B Rank'}, inplace=True)
df_lrmcB.rename(columns={'All games (vs. Div. I opponents)': 'Team'}, inplace=True)
df_lrmcB = df_lrmcB[['Team', 'LRMC B Rank']]
```

```python
df_lrmcB = df_lrmcB.dropna(subset=['Team'])
df_lrmcB = df_lrmcB[df_lrmcB['Team'] != 'Team']
df_lrmcB['Team'] = df_lrmcB['Team'].str.replace('_St', '_St.', regex=False)
df_lrmcB['Team'] = df_lrmcB['Team'].str.replace('_', ' ')

df_lrmcB['LRMC B Rank'] = df_lrmcB['LRMC B Rank'].astype(float)

display(df_lrmcB.head(5))
display(df_lrmcB.info())
```

```python
# Perform a left join on "Team"
df_lrmcB = df_lrmcB.merge(df_teamsIndex, on='Team', how='left')

df_lrmcB = df_lrmcB.drop(columns=['Unnamed: 0'])

# Display the first few rows to verify
print(df_lrmcB.head(5))
```

```python
# Convert to csv

df_lrmcB.to_csv("../L1/data/lrmcB_L1.csv")
```
