```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_vegas = pd.read_csv('data/vegasOdds_L0.csv')

display(df_vegas.head(5))
display(df_vegas.info())
```

```python
df_vegas = df_vegas.drop(columns=["Rank", "Odds"])

df_vegas["Implied Probability"] = df_vegas["Implied Probability"].str.rstrip('%').astype(float) / 100
```

```python
df_vegas["Team"] = df_vegas["Team"].str.replace(r"\bState\b", "St.", regex=True)
```

```python
df_vegas = df_vegas.rename(columns={"Implied Probability": "Vegas_Prob"})
```
