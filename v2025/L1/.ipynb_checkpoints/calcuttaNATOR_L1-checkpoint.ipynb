```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown
```

```python
df_calcuttaNATOR = pd.read_csv('data/calcuttaNATOR_L1.csv')
df_actualsSmoother = pd.read_csv('../L2/data/actualsSmoother_L2.csv')
df_calcuttaValueHistory = pd.read_csv('data/calcuttaValueHistory_L1.csv')
df_drafted = pd.read_csv('data/drafted_L1.csv')
df_VEGsemble = pd.read_csv("../L2/data/barttorvik_VEGsemblePredict2025_L2.csv")

display(df_calcuttaNATOR.head(5))
display(df_calcuttaNATOR.info())

display(df_actualsSmoother.info())
display(df_calcuttaValueHistory.info())

display(df_drafted.head(5))
display(df_drafted.info())

display(df_VEGsemble.head(5))
display(df_VEGsemble.info())
```

```python
df_calcuttaNATOR = df_calcuttaNATOR.drop(columns=['Unnamed: 0'])
df_actualsSmoother = df_actualsSmoother.drop(columns=['Unnamed: 0'])
df_calcuttaValueHistory = df_calcuttaValueHistory.drop(columns=['Unnamed: 0'])
df_VEGsemble = df_VEGsemble.drop(columns=['Unnamed: 0'])
```

```python
# Compute "Avg" as the weighted average of "Golden Ratio" and "VEGsemble"
df_calcuttaNATOR['Avg'] = (df_calcuttaNATOR['Golden Ratio'] * 0.4) + (df_calcuttaNATOR['VEGsemble'] * 0.6)



```

```python
df_calcuttaNATOR = df_calcuttaNATOR.merge(df_actualsSmoother, on='Seed', how='left')
#df_goldenRatio = df_goldenRatio.merge(df_kenPom, on='Team', how='left')
#df_goldenRatio = df_goldenRatio.merge(df_espnBPI, on='Team', how='left')
#df_goldenRatio = df_goldenRatio.merge(df_lrmcB, on='Team', how='left')
#df_goldenRatio = df_goldenRatio.merge(df_lrmcC, on='Team', how='left')

display(df_calcuttaNATOR.head(5))
display(df_calcuttaNATOR.info())
```

```python
# Replace NaN values in "smoothPtsT" with 0
df_calcuttaNATOR['smoothPtsT'] = df_calcuttaNATOR['smoothPtsT'].fillna(0)
```

```python
# Compute "AdjAvg" as the sum of "Avg" and "smoothPtsT"
df_calcuttaNATOR['AdjAvg'] = df_calcuttaNATOR['Avg'] + df_calcuttaNATOR['smoothPtsT']

# Display the first few rows
display(df_calcuttaNATOR.head())
```

```python
# Compute the sum of the "Avg" column
sumAvg = df_calcuttaNATOR['Avg'].sum()
# Display the result
print("sumAvg:", sumAvg)

# Compute the sum of the "Avg" column
sumAdjAvg = df_calcuttaNATOR['AdjAvg'].sum()
# Display the result
print("sumAdjAvg:", sumAdjAvg)

# Compute the average of the "Total Points" column
avgHistoryTotalPts = df_calcuttaValueHistory['Total Points'].mean()
# Display the result
print("avgHistoryTotalPts:", avgHistoryTotalPts)
```

```python
# Compute "histAdjAvg" based on the given formula
#df_calcuttaNATOR['histAdjAvg'] = ((((avgHistoryTotalPts - sumAvg) / sumAvg) + 1) * df_calcuttaNATOR['AdjAvg'])
#df_calcuttaNATOR['histAdjAvg'] = ((((avgHistoryTotalPts - sumAvg) / sumAvg) + 1) * df_calcuttaNATOR['AdjAvg'])

historicalRatio = avgHistoryTotalPts / sumAdjAvg
df_calcuttaNATOR['histAdjAvg'] =  historicalRatio * df_calcuttaNATOR['AdjAvg']

# Compute the sum of the "Avg" column
sumHistAdjAvg = df_calcuttaNATOR['histAdjAvg'].sum()

# Display the result
print("histAdjAvg:", sumHistAdjAvg)

# Display the first few rows
display(df_calcuttaNATOR.head())
```

```python
# Assign "Floor" as "histAdjAvg"
df_calcuttaNATOR['Floor'] = df_calcuttaNATOR['histAdjAvg']

# Assign "Ceiling" as "histAdjAvg" * 2
df_calcuttaNATOR['Ceiling'] = df_calcuttaNATOR['histAdjAvg'] * 2

# Display the first few rows
display(df_calcuttaNATOR.head())
```

```python
# Ensure df_VEGsemble has the necessary column and merge on "Team"
if 'VEGsemble_Prob' in df_VEGsemble.columns:
    df_calcuttaNATOR = df_calcuttaNATOR.merge(df_VEGsemble[['Team', 'VEGsemble_Prob']], on='Team', how='left')

# Create "VEGsemble Q" with quartiles, ensuring Q1 is the highest quartile
df_calcuttaNATOR['VEGsemble Q'] = pd.qcut(
    df_calcuttaNATOR['VEGsemble_Prob'], 
    q=8, 
    labels=['Q8', 'Q7', 'Q6', 'Q5','Q4','Q3', 'Q2', 'Q1']  # Highest values get "Q1"
)

# Create a new column for the rank of "VEGsemble_Prob" in df_calcuttaNATOR
df_calcuttaNATOR['VEGsemble_Prob Rank'] = df_calcuttaNATOR['VEGsemble_Prob'].rank(ascending=False, method='min')


# Define the desired column order
new_column_order = [
    'Seed', 'Team', 'VEGsemble_Prob Rank', 'VEGsemble_Prob', 'VEGsemble Q', 'Golden Ratio', 'VEGsemble', 
    'Avg', 'smoothPtsT', 'AdjAvg', 'histAdjAvg', 'Floor', 'Ceiling'
]

# Reorder the columns in df_calcuttaNATOR
df_calcuttaNATOR = df_calcuttaNATOR[new_column_order]

# Display the first few rows
display(df_calcuttaNATOR.head(5))
```

```python
# Filter rows where "VEGsemble_Prob" is NaN in df_calcuttaNATOR
df_calcuttaNATOR_nan_vegsemble = df_calcuttaNATOR[df_calcuttaNATOR['VEGsemble_Prob'].isna()]

# Display the first few rows
display(df_calcuttaNATOR_nan_vegsemble)
```

```python
Q="Q1"

# Filter rows where "VEGsemble Q" is "Q1" in df_calcuttaNATOR
df_calcuttaNATOR_Q = df_calcuttaNATOR[df_calcuttaNATOR['VEGsemble Q'] == Q]

# Display the first few rows
display(df_calcuttaNATOR_Q)
```
