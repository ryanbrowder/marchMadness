```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_actualsSmoother = pd.read_csv('data/actualsSmoother_L1.csv')

display(df_actualsSmoother.head(5))
display(df_actualsSmoother.info())
```

```python
df_actualsSmoother = df_actualsSmoother.drop(columns=['Unnamed: 0'])
display(df_actualsSmoother.info())
```

```python
# Group by "Seed" and calculate sum of "Projected" and "Actual", and count occurrences
df_seed_ratio = df_actualsSmoother.groupby('Seed').agg({'Projected': 'sum', 'Actual': 'sum', 'Seed': 'count'}).rename(columns={'Seed': 'n'}).reset_index()

# Compute the new metric: (Actual - Projected) / n and rename it to "smoothPtsT"
df_seed_ratio['smoothPtsT'] = (df_seed_ratio['Actual'] - df_seed_ratio['Projected']) / df_seed_ratio['n']

# Display the resulting DataFrame
print(df_seed_ratio.head())  # Print first few rows
```

```python
df_seed_ratio = df_seed_ratio[['Seed', 'smoothPtsT']]

# Display the resulting DataFrame
print(df_seed_ratio.head(20))  # Print first few rows
```

```python
# Convert to csv

df_seed_ratio.to_csv("../L2/data/actualsSmoother_L2.csv")
```
