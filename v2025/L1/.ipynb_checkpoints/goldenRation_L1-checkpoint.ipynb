```python
# import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df_goldenRatio = pd.read_csv('data/tourneyTeams_L1.csv')
df_powerRank = pd.read_csv('data/powerRank_L1.csv')
df_kenPom = pd.read_csv('data/kenPom_L1.csv')
df_espnBPI = pd.read_csv('data/espnBPI_L1.csv')
df_lrmcB = pd.read_csv('data/lrmcB_L1.csv')
df_lrmcC = pd.read_csv('data/lrmcC_L1.csv')

display(df_goldenRatio.head(5))
display(df_goldenRatio.info())

display(df_powerRank.info())
display(df_kenPom.info())
display(df_espnBPI.info())
display(df_lrmcB.info())
display(df_lrmcC.info())
```

```python
df_goldenRatio = df_goldenRatio.drop(columns=['Unnamed: 0'])
df_powerRank = df_powerRank.drop(columns=['Unnamed: 0'])
df_kenPom = df_kenPom.drop(columns=['Unnamed: 0'])
df_espnBPI = df_espnBPI.drop(columns=['Unnamed: 0'])
df_lrmcB = df_lrmcB.drop(columns=['Unnamed: 0'])
df_lrmcC = df_lrmcC.drop(columns=['Unnamed: 0'])
```

```python
df_goldenRatio = df_goldenRatio.merge(df_powerRank, on='Index', how='left')
df_goldenRatio = df_goldenRatio.merge(df_kenPom, on='Index', how='left')
df_goldenRatio = df_goldenRatio.merge(df_lrmcB, on='Index', how='left', suffixes=('', '_lrmcB'))
df_goldenRatio = df_goldenRatio.merge(df_espnBPI, on='Index', how='left', suffixes=('', '_espnBPI'))
df_goldenRatio = df_goldenRatio.merge(df_lrmcC, on='Index', how='left', suffixes=('', '_lrmcC'))

display(df_goldenRatio.head(50))
display(df_goldenRatio.info())
```

```python
df_goldenRatio['Golden Ratio'] = None  # Initialize with NaN or any default value

# Define the weights for each ranking system
weights = {
    'PowerRank Rank': 0.3865,
    'KenPom Rank': 0.3865,
    'BPI Rank': 0.1534,
    'LRMC B Rank': 0.0368,
    'LRMC C Rank': 0.0368
}

# Compute the weighted sum of available rankings (ignoring NaNs)
df_goldenRatio['Golden Ratio'] = df_goldenRatio[list(weights.keys())].mul(weights).sum(axis=1) / df_goldenRatio[list(weights.keys())].notna().mul(weights).sum(axis=1)

df_goldenRatio = df_goldenRatio.sort_values(by='Team').reset_index(drop=True)

# Display the result
display(df_goldenRatio.head())
```

```python
# Drop the unwanted columns related to duplicate "Team" entries
df_goldenRatio = df_goldenRatio.drop(columns=['Team','Team_y','Team_lrmcB', 'Team_espnBPI', 'Team_lrmcC'], errors='ignore')

# Reorder columns to make "Team" the second column
columns = list(df_goldenRatio.columns)
columns.remove('Team_x')  # Remove "Team" from its current position
columns.insert(1, 'Team_x')  # Insert "Team" as the second column

# Reassign DataFrame with the new column order
df_goldenRatio = df_goldenRatio[columns]

# Rename "Team_x" to "Team"
df_goldenRatio.rename(columns={'Team_x': 'Team'}, inplace=True)

# Display the updated DataFrame
display(df_goldenRatio.head(100))
```

```python
# Filter rows where "PowerRank Rank" is NaN in df_goldenRatio
df_goldenRatio_nan_Rank = df_goldenRatio[df_goldenRatio['PowerRank Rank'].isna()]
display(df_goldenRatio_nan_Rank)
```
